```
      # this prints the number of queries for which we can find the answer in the context alongwith total number of queries
count = 0
total_count = 0
for _document in range(0,len(data['data'])):
  document = data['data'][_document]
  for _paragraph in range(0,len(document['paragraphs'])):
    paragraph = document['paragraphs'][_paragraph]
    for _qa in paragraph['qas']:
      total_count = total_count + 1
      if(_qa['is_impossible'] == True):
        count = count+1
        
print(count, total_count)
```
